```python
# borrow from makeittalk
import ipywidgets as widgets
import glob
import matplotlib.pyplot as plt
print("Choose the image name to animate: (saved in folder 'examples/')")
img_list = glob.glob1('examples/source_image', '*.png')
img_list.sort()
img_list = [item.split('.')[0] for item in img_list]
default_head_name = widgets.Dropdown(options=img_list, value='art_12')
def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
        plt.imshow(plt.imread('examples/source_image/{}.png'.format(default_head_name.value)))
        plt.axis('off')
        plt.show()
default_head_name.observe(on_change)
display(default_head_name)
plt.imshow(plt.imread('examples/source_image/{}.png'.format(default_head_name.value)))
plt.axis('off')
plt.show()
```

```python
# Run this code cell for uploading image and audio

%%writefile app.py
import streamlit as st
import os
import subprocess
import glob

# Function to generate animation
def generate_animation(img_path, audio_path):
    # Run inference to generate animation
    subprocess.run([
        "python3.8", "inference.py",
        "--driven_audio", audio_path,
        "--source_image", img_path,
        "--result_dir", "./results",
        "--still", "--preprocess", "full",
        "--enhancer", "gfpgan"
    ])

# Main Streamlit app
def main():
    st.title("Real-Time Talking-Head Animation")

    # Upload image
    st.subheader("1. Upload Image")
    uploaded_image = st.file_uploader("Upload Image", type=["jpg", "jpeg", "png"])

    # Upload audio
    st.subheader("2. Upload Audio")
    uploaded_audio = st.file_uploader("Upload Audio", type=["wav"])

    # Generate animation
    if st.button("Generate Animation"):
        if uploaded_image is None or uploaded_audio is None:
            st.error("Please upload both an image and an audio file.")
        else:
            with st.spinner("Generating Animation..."):
                # Save uploaded files
                img_path = "uploaded_image.png"
                audio_path = "uploaded_audio.wav"
                with open(img_path, "wb") as f:
                    f.write(uploaded_image.read())
                with open(audio_path, "wb") as f:
                    f.write(uploaded_audio.read())

                # Run inference to generate animation
                generate_animation(img_path, audio_path)
            st.success("Animation generated successfully!")

            # Display latest generated animation video
            st.subheader("Latest Generated Animation Video")
            results = sorted(glob.glob('./results/*.mp4'))
            if results:
                latest_animation = results[-1]
                video_file = open(latest_animation, 'rb')
                video_bytes = video_file.read()
                st.video(video_bytes)
            else:
                st.info("No animation video available.")

if __name__ == "__main__":
    main()
```
